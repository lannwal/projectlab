```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import itertools
from scipy.stats import pearsonr,ttest_ind

import statsmodels.api as sm
from statsmodels.formula.api import ols 
```

test anova from statsmodels

```python
moore = sm.datasets.get_rdataset("Moore", "carData",
                                    cache=True) # load data
data = moore.data

#print(data)
data = data.rename(columns={"partner.status":
                                "partner_status"}) # make name pythonic
#print(data)
moore_lm = ols('conformity ~ partner_status * C(fcategory) + C(fscore)', data=data).fit()
   
table = sm.stats.anova_lm(moore_lm, typ=2) # Type 2 ANOVA DataFrame
print(table)
# 'gene_X_sensitivity ~MSI_status + C(Tissue) + driver_gene_Y_status'
```
